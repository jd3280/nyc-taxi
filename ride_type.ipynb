行程类型与价格的关系

```python
import pandas as pd
df = pd.read_csv('/Users/dingjiadong/Desktop/uber/cab_rides.csv')
#df.head()
```

```python
df['name'].unique()
```

```python
share = df[(df['name']=='Shared') | (df['name']=='UberPool')]
#share
```

```python
normal = df[(df['name']=='Lyft')]
#normal
```

```python
lux = df[(df['name']=='Lux') | (df['name']=='Black') | (df['name']=='Black SUV') | (df['name']=='Lux Black')]
#lux
```

```python
large = df[(df['name']=='Lux Black XL') | (df['name']=='Lyft XL') | (df['name']=='UberXL') | (df['name']=='UberX')
| (df['name']=='WAV')]
#large
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
```

```python
df_type = pd.DataFrame({'share':share.price,'normal':normal.price,'lux':lux.price,'large':large.price,})
df_type.boxplot()
plt.xlabel('乘坐类型')
plt.title('乘坐类型与价格比较')
plt.show()
```

以上为不同乘坐类型的价格分布图，可以明显看出拼车的价格上界、众数、以及50%的价格分布都是最低的，普通乘坐类型的价格分布稍高于拼车，豪华车的价格下界、众数、以及50%的价格分布是最高的，大型车的价格分布仅此于豪华车。这个结果符合我们的预期，越豪华的乘坐类型，价格会越高。
