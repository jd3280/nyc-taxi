#早晨路线

```python
import pandas as pd
df = pd.read_csv('/Users/dingjiadong/Desktop/uber/cab_rides.csv')
df['time_stamp'] = pd.to_datetime(df['time_stamp'], unit='ms')
df_path = df[['time_stamp','destination','source']]
df_path['time_stamp'] = df_path['time_stamp'].dt.hour###将时间戳转为小时
```

```python
#df
```

```python
df_mrn = df_path[(df_path['time_stamp'] >=6) & (df_path['time_stamp'] <=9)]
#df_mrn
```

```python
#df_mrn['frequency'] = 1
#df_mrn.head(20)
```

```python
df_mrn.groupby(by=['source','destination'], as_index=False).count().sort_values(by='time_stamp',ascending=False)
```

早晨最热门的路线是Beacon Hill到Northodox End,其次是Financial District到Haymarket Square 然后是Back Bay到North End

```python
df_mrn.destination.value_counts()
```

```python
df_mrn.source.value_counts()
```

最热门的出发地依次是Financial District，Back Bay，Beacon Hill；最热门的目的地依次是South Station，North End, West End. 

#傍晚路线

```python
df_eve = df_path[(df_path['time_stamp'] >=16) & (df_path['time_stamp'] <=18)] #傍晚
```

```python
df_eve.source.value_counts()
```

```python
df_eve.destination.value_counts()
```

在傍晚时，最热门的出发地是Fenway，Theatre District，Financial District；最热门的目的地是Financial District, Theatre District, Northeastern University.

```python
df_eve.groupby(by=['source','destination'], as_index=False).count().sort_values(by='time_stamp',ascending=False)
```

常见的迁移路线是Financial District到Haymarket Square, North End到Back Bay, Boston Uni到Theatre District

#工作日与周末

```python
df = pd.read_csv('/Users/dingjiadong/Desktop/uber/任务10数据集/cab_rides.csv')
df['time_stamp'] = pd.to_datetime(df['time_stamp'], unit='ms')
df_wkd = df
df_wkd['time_stamp'] = df_wkd['time_stamp'].dt.weekday
```

```python
df_wday = df_wkd[df_wkd.time_stamp < 5]
df_wend = df_wkd[df_wkd.time_stamp > 4]
```

```python
df_wend.head()
```

```python
df_wend.destination.value_counts()
```

```python
df_wday.destination.value_counts()
```

周末的热门目的地是West End，Fenway，Financial District；工作日的热门目的地为Financial District，Back Bay，Theatre District

```python
df_wend.source.value_counts()
```

```python
df_wday.source.value_counts()
```

周末的热门出发地是Financial District，Haymarket Square，Back Bay；工作日的热门出发地的地为Financial District，South Station，Theatre District

综上，司机可以前往在早晨最繁忙的Financial District等地接单；在傍晚，热门的目的地和出发地都包括了Theatre District和Financial District，司机可以通过前往这些地方接单。工作日与周末的热门出发地中都包括了Financial District，可以注意这里的情况，其他目的地则可以按照工作日与否调整接单地点。
